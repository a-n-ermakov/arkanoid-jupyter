```python
%pylab

import math
import random
from matplotlib import pyplot as plt
from matplotlib import animation

fig = plt.figure()
fig.set_dpi(100)
fig.set_size_inches(7, 6.5)

ax = plt.axes(xlim=(0, 10), ylim=(0, 10))
ball = plt.Circle((5, -5), 0.2, fc='y')
rack = plt.Rectangle((4, 0.3), 2, 0.2, fc='b')


def init_ball(patch):
    patch.center = tuple(5 + np.random.rand(2))
    patch.respawn = 0
    v_abs = .15
    alpha = math.pi/8 + math.pi/4 * random.random() + math.pi/2 * random.randint(0, 1)
    patch.v = (v_abs * math.cos(alpha), v_abs * math.sin(alpha))

    
def init_rack(patch):
    patch.v = 0

    
def init():
    init_ball(ball)
    ax.add_patch(ball)
    init_rack(rack)
    ax.add_patch(rack)
    return ball, rack


def animate_ball(patch):
    if patch.respawn > 0:
        patch.respawn += 1
        if patch.respawn > 100:
            patch.respawn = 0
            init_ball(patch)
        else:    
            return
    
    x, y = patch.center
    r = patch.radius
    dx, dy = patch.v
    left, right = ax.get_xlim()
    bottom, top = ax.get_ylim()
    bottom = rack.get_y() + rack.get_height()
    x += dx
    # left and right reflect
    if x - r < left or x + r > right:
        dx = -dx
        x += dx
    y += dy
    # top reflect
    if y + r > top:
        dy = -dy
        y += dy
    # rack reflect    
    if y - r < bottom:
        if x > rack.get_x() and x < rack.get_x() + rack.get_width():
            dy = -dy
            y += dy
        else:
            patch.respawn += 1
        
    patch.center = (x, y)
    patch.v = (dx, dy)

    
def animate_rack(patch):
    left, right = ax.get_xlim()
    x, w = patch.get_x(), patch.get_width()
    dx = patch.v
    if dx != 0:
        x += 0.1 * dx
        if x > left and x < right - w:
            print('-' if dx < 0 else '+', end='')
            patch.set_x(x)


def animate(i):
    animate_ball(ball)
    animate_rack(rack)
    return ball, rack

anim = animation.FuncAnimation(fig, animate, 
                               init_func=init, 
                               frames=36000, 
                               interval=20,
                               blit=True)

# plt.show()

def onclick(event):
    print('%s click: button=%d, x=%d, y=%d, xdata=%f, ydata=%f' %
          ('double' if event.dblclick else 'single', event.button,
           event.x, event.y, event.xdata, event.ydata))

def onpress(event):
#     print('key press: key=%s, x=%d, y=%d, xdata=%f, ydata=%f' %
#           (event.key, event.x, event.y, event.xdata, event.ydata))
    global rack
    if event.key == ',': 
        rack.v = -1
#         print(rack.v)
    elif event.key == '.':
        rack.v = 1
#         print(rack.v)

def onrelease(event):
#     print('key press: key=%s, x=%d, y=%d, xdata=%f, ydata=%f' %
#           (event.key, event.x, event.y, event.xdata, event.ydata))
    global rack
    if event.key == ',' or event.key == '.':
        rack.v = 0
#         print('release ', rack.v)

cid1 = fig.canvas.mpl_connect('button_press_event', onclick)
cid2 = fig.canvas.mpl_connect('key_press_event', onpress)
cid3 = fig.canvas.mpl_connect('key_release_event', onrelease)
```
